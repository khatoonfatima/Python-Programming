Tuples are ordered collection of items that are immutable. They are similar to lists, but their immutability makes them different.

```python
#Packing and unpacking tuples
#Packing
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
#Unpacking
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
#Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
nested_tuple=((1,2,3),("a","b","c"),(True,False))
#Access the elements inside a tuple
print(nested_tuple[0])
```

```python
print(nested_tuple[1])
```

```python
print(nested_tuple[1][2]) #For displaying C from the list
```

```python
##Iterating over nested tuples
for sub_tuple in nested_tuple: #Sub_tuple is the tuple within the tuple
  for item in sub_tuple:       #Taking the item from the sub_tuple
      print(item,end="")       #Printing the item from the sub_tuple
  print()

```
